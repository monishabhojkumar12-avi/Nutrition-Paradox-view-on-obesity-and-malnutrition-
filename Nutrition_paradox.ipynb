```python
import pandas as pd

obesity_adults = pd.read_csv("/content/obesity.csv")
obesity_children = pd.read_csv("/content/obesity2.csv")

malnutrition_adults = pd.read_csv("/content/malnutrition.csv")
malnutrition_children = pd.read_csv("/content/malnutrition2.csv")
```

```python
obesity_adults["age_group"] = "Adults"
obesity_children["age_group"] = "Children"
malnutrition_adults["age_group"] = "Adults"
malnutrition_children["age_group"] = "Children"
df_obesity = pd.concat(
    [obesity_adults, obesity_children],
    ignore_index=True
)
df_malnutrition = pd.concat(
    [malnutrition_adults, malnutrition_children],
    ignore_index=True
)

df_obesity = df_obesity[
    (df_obesity["TimeDim"] >= 2012) &
    (df_obesity["TimeDim"] <= 2022)
]

df_malnutrition = df_malnutrition[
    (df_malnutrition["TimeDim"] >= 2012) &
    (df_malnutrition["TimeDim"] <= 2022)
]
```

```python
df_obesity.head()
df_malnutrition.head()

#df_obesity["age_group"].value_counts()
#df_malnutrition["age_group"].value_counts()
```

```python
columns_to_keep = [
    "ParentLocation",
    "Dim1",
    "TimeDim",
    "Low",
    "High",
    "NumericValue",
    "SpatialDim",
    "age_group"
]

df_obesity = df_obesity[columns_to_keep]
df_malnutrition = df_malnutrition[columns_to_keep]

rename_columns = {
    "TimeDim": "Year",
    "Dim1": "Gender",
    "NumericValue": "Mean_Estimate",
    "Low": "LowerBound",
    "High": "UpperBound",
    "ParentLocation": "Region",
    "SpatialDim": "Country"
}

df_obesity.rename(columns=rename_columns, inplace=True)
df_malnutrition.rename(columns=rename_columns, inplace=True)

```

```python
def standardize_gender(value):
    if value is None:
        return value
    value = value.lower()
    if "male" in value:
        return "Male"
    elif "female" in value:
        return "Female"
    else:
        return "Both"

df_obesity["Gender"] = df_obesity["Gender"].apply(standardize_gender)
df_malnutrition["Gender"] = df_malnutrition["Gender"].apply(standardize_gender)
```

```python
import pycountry
```

```python
def alpha3_to_country_name(code):
    try:
        country = pycountry.countries.get(alpha_3=code)
        if country:
            return country.name
        else:
            return None
    except:
        return None
```

```python
df_obesity["Country"] = df_obesity["Country"].apply(alpha3_to_country_name)
df_malnutrition["Country"] = df_malnutrition["Country"].apply(alpha3_to_country_name)
```

```python
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}
```

```python
def convert_country(code):
    if code in special_cases:
        return special_cases[code]
    try:
        country = pycountry.countries.get(alpha_3=code)
        if country:
            return country.name
    except:
        pass
    return code
```

```python
df_obesity["Country"] = df_obesity["Country"].apply(convert_country)
df_malnutrition["Country"] = df_malnutrition["Country"].apply(convert_country)
```

```python
df_obesity["Age_Group"] = df_obesity["Mean_Estimate"].apply(
    lambda x: "Adult" if x >= 18 else "Child/Adolescent"
)
df_malnutrition["Age_Group"] = df_malnutrition["Mean_Estimate"].apply(
    lambda x: "Adult" if x >= 18 else "Child/Adolescent"
)
```

```python
df_obesity["CI_Width"] = df_obesity["UpperBound"] - df_obesity["LowerBound"]
df_malnutrition["CI_Width"] = df_malnutrition["UpperBound"] - df_malnutrition["LowerBound"]
```

```python
def obesity_category(value):
    if value >= 30:
        return "High"
    elif value >= 25:
        return "Moderate"
    else:
        return "Low"
df_obesity["Obesity_Level"] = df_obesity["Mean_Estimate"].apply(obesity_category)
```

```python
def malnutrition_category(value):
    if value >= 20:
        return "High"
    elif value >= 10:
        return "Moderate"
    else:
        return "Low"
df_malnutrition["Malnutrition_Level"] = df_malnutrition["Mean_Estimate"].apply(malnutrition_category)
```

```python
#10.Countries where female obesity exceeds male by large margin (same year)
query = """
SELECT
    f.Country,
    f.Year,
    f.Mean_Estimate AS female_obesity,
    m.Mean_Estimate AS male_obesity,
    ROUND(f.Mean_Estimate - m.Mean_Estimate, 2) AS obesity_gap
FROM obesity f
JOIN obesity m
  ON f.Country = m.Country
 AND f.Year = m.Year
WHERE f.Gender = 'Female'
  AND m.Gender = 'Male'
  AND f.Mean_Estimate IS NOT NULL
  AND m.Mean_Estimate IS NOT NULL
  AND (f.Mean_Estimate - m.Mean_Estimate) >= 10
ORDER BY obesity_gap DESC;
"""


female_male_gap = pd.read_sql(query, conn)
female_male_gap.head()
```

```python
#Join Query
#1.Obesity vs malnutrition comparison by country(any 5 countries)
query = """
SELECT o.Country,
       AVG(o.Mean_Estimate) AS avg_obesity,
       AVG(m.Mean_Estimate) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
ON o.Country = m.Country
WHERE o.Country IN ('India','USA','Brazil','Nigeria','China')
GROUP BY o.Country;
"""
obesity_malnutrition_comparison = pd.read_sql(query, conn)
print(obesity_malnutrition_comparison)
```

```python
#2.Gender-based disparity in both obesity and malnutrition
query = """
SELECT o.Gender,
       AVG(o.Mean_Estimate) AS avg_obesity,
       AVG(m.Mean_Estimate) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
ON o.Country = m.Country AND o.Gender = m.Gender
GROUP BY o.Gender;
"""
gender_comparison = pd.read_sql(query, conn)
print(gender_comparison)
```

```python
#3.Region-wise avg estimates side-by-side(Africa and America)
query = """
SELECT o.Region,
       AVG(o.Mean_Estimate) AS avg_obesity,
       AVG(m.Mean_Estimate) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
ON o.Country = m.Country
WHERE o.Region IN ('Africa','America')
GROUP BY o.Region;
"""
region_comparison = pd.read_sql(query, conn)
print(region_comparison)
```

```python
#4.Countries with obesity up & malnutrition down
query = """
WITH obesity_years AS (
    SELECT Country,
           MIN(Year) AS start_year,
           MAX(Year) AS end_year
    FROM obesity
    GROUP BY Country
),
obesity_trend AS (
    SELECT
        o.Country,
        sy.start_year,
        sy.end_year,
        MAX(CASE WHEN o.Year = sy.start_year THEN o.Mean_Estimate END) AS start_obesity,
        MAX(CASE WHEN o.Year = sy.end_year THEN o.Mean_Estimate END) AS end_obesity
    FROM obesity o
    JOIN obesity_years sy
      ON o.Country = sy.Country
    GROUP BY o.Country
),
malnutrition_years AS (
    SELECT Country,
           MIN(Year) AS start_year,
           MAX(Year) AS end_year
    FROM malnutrition
    GROUP BY Country
),
malnutrition_trend AS (
    SELECT
        m.Country,
        my.start_year,
        my.end_year,
        MAX(CASE WHEN m.Year = my.start_year THEN m.Mean_Estimate END) AS start_malnutrition,
        MAX(CASE WHEN m.Year = my.end_year THEN m.Mean_Estimate END) AS end_malnutrition
    FROM malnutrition m
    JOIN malnutrition_years my
      ON m.Country = my.Country
    GROUP BY m.Country
)
SELECT o.Country,
       o.start_obesity,
       o.end_obesity,
       m.start_malnutrition,
       m.end_malnutrition
FROM obesity_trend o
JOIN malnutrition_trend m
  ON o.Country = m.Country
WHERE o.end_obesity > o.start_obesity
  AND m.end_malnutrition < m.start_malnutrition;
"""

country_comparison = pd.read_sql(query, conn)
print(country_comparison)

```

```python
#5.Age-wise trend analysis
query = """
SELECT
    o.Year,
    o.Age_Group,
    AVG(o.Mean_Estimate) AS avg_obesity,
    AVG(m.Mean_Estimate) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
  ON o.Country = m.Country
 AND o.Year = m.Year
 AND o.Age_Group = m.Age_Group
GROUP BY o.Year, o.Age_Group
ORDER BY o.Year, o.Age_Group;
"""
age_group_comparison = pd.read_sql(query, conn)
print(age_group_comparison)
```

```python
#5.Age groups with rising obesity & falling malnutrition
query = """
WITH obesity_years AS (
    SELECT Age_Group,
           MIN(Year) AS start_year,
           MAX(Year) AS end_year
    FROM obesity
    GROUP BY Age_Group
),
obesity_trend AS (
    SELECT
        o.Age_Group,
        MAX(CASE WHEN o.Year = oy.start_year THEN o.Mean_Estimate END) AS start_obesity,
        MAX(CASE WHEN o.Year = oy.end_year THEN o.Mean_Estimate END) AS end_obesity
    FROM obesity o
    JOIN obesity_years oy
      ON o.Age_Group = oy.Age_Group
    GROUP BY o.Age_Group
),
malnutrition_years AS (
    SELECT Age_Group,
           MIN(Year) AS start_year,
           MAX(Year) AS end_year
    FROM malnutrition
    GROUP BY Age_Group
),
malnutrition_trend AS (
    SELECT
        m.Age_Group,
        MAX(CASE WHEN m.Year = my.start_year THEN m.Mean_Estimate END) AS start_malnutrition,
        MAX(CASE WHEN m.Year = my.end_year THEN m.Mean_Estimate END) AS end_malnutrition
    FROM malnutrition m
    JOIN malnutrition_years my
      ON m.Age_Group = my.Age_Group
    GROUP BY m.Age_Group
)
SELECT o.Age_Group,
       o.start_obesity,
       o.end_obesity,
       m.start_malnutrition,
       m.end_malnutrition
FROM obesity_trend o
JOIN malnutrition_trend m
  ON o.Age_Group = m.Age_Group
WHERE o.end_obesity > o.start_obesity
  AND m.end_malnutrition < m.start_malnutrition;
"""

age_group_comparison = pd.read_sql(query, conn)
print(age_group_comparison)
```
